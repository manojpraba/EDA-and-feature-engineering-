```python
import pandas as pd

# Assuming you have two DataFrames with a common column "Country Code"
df1 = pd.DataFrame({
    'Country Code': ['US', 'CA', 'IN', 'UK'],
    'Population': [327, 38, 1380, 67]
})

df2 = pd.DataFrame({
    'Country Code': ['US', 'CA', 'IN', 'AU'],
    'GDP (Billion $)': [21.43, 1.84, 2.87, 1.39]
})

# Merge DataFrames based on the "Country Code" column
merged_df = pd.merge(df1, df2, on='Country Code', how='right')

# Display the merged DataFrame
print(merged_df)
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'count'})
```

```python
ratings
```

## Observation
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```
